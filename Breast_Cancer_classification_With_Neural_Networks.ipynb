**IMPORTING THE DEPENDENCIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

DATA COLLECTION

```python
bdata = sklearn.datasets.load_breast_cancer()
```

```python
print(bdata)
```

```python
data = pd.DataFrame(bdata.data, columns = bdata.feature_names)
```

```python
data['label'] = bdata.target
```

seperating the feautures and target

```python
x = data.drop(columns = 'label',axis=1)
y = data['label']
```

```python
print(x)
```

```python
print(y)
```

train_test splittting

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

standardisation the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

**building the neural networks**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
from keras.src.layers.core.dense import Dense
#setting the layers
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation ='relu'),
                          keras.layers.Dense(2, activation ='sigmoid'),
])
```

```python
#compiling the neural network

model.compile(optimizer='adam',
              loss ='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

training

```python
history= model.fit(x_train_std,y_train, validation_split=0.1, epochs = 25)
```

visualisation

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy on test data

```python
loss, accuracy = model.evaluate(x_test_std,y_test)
print(accuracy)
```
